## Corrigindo as colunas

```python
#Esta funÃ§Ã£o compara as colunas entre dois DataFrames ('original_df' e 'df1') e retorna uma lista das colunas presentes em 'df1' que nÃ£o estÃ£o presentes em 'original_df'.
# Isso ajuda a identificar quais colunas foram adicionadas ou estÃ£o ausentes em relaÃ§Ã£o ao DataFrame original.
def check_errors(original_df, df1):
    original_columns = original_df.columns
    columns1 = df1.columns
    colunas = []
    for i in columns1:
      if i not in original_columns:
        colunas.append(i)

    return colunas
```

```python
#utilizando a funÃ§Ã£o para comparaÃ§Ã£o dos anos 2017 e 2018
check_errors(happiness_2017, happiness_2018)
```
